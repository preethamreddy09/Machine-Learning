```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('carseats.csv')
df['Sales']= pd.cut(x=df['Sales'], bins=[0, 3, 10.5,16.3],
                     labels=['low', 'medium', 'high'])
df.head()
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,:1]
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x['US'] = le.fit_transform(x['US'])
x['Urban'] = le.fit_transform(x['Urban'])
x.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
en_x = pd.DataFrame(ohe.fit_transform(x[['ShelveLoc']]).toarray())
x = x.join(en_x)
x.head()
```

```python
d = x.drop('ShelveLoc',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
ss = StandardScaler()
x_train.iloc[:,0:6]= ss.fit_transform(x_train.iloc[:,0:6])
x_test.iloc[:,0:6] = ss.fit_transform(x_test.iloc[:,0:6])
#x_train= ss.fit_transform(x_train)
#x_test = ss.fit_transform(x_test)
```

```python
x_train.drop([174],axis=0,inplace=True)
y_train.drop([174],axis=0,inplace=True)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics
```

```python
x_train.dtypes
```

```python
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
ac
```
